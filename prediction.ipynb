### Load the Model and Get the Results

```python
import torch
from MolNexTR import molnextr
IMAGE = './examples/1.png'
MODEL = './checkpoints/molnextr_best.pth'
device = torch.device('cpu')
model = molnextr(MODEL, device)
predictions = model.predict_final_results(IMAGE, return_atoms_bonds=True)
print(predictions)
smiles = predictions['predicted_smiles']
molfile = predictions['predicted_molfile']
print('MolFile:')
print(molfile)
print('SMILES:')
print(smiles)
```
